# 1. Estruturas de Dados em R

Nesta sessão, vamos aprofundar os diferentes tipos de estruturas de dados que o R utiliza para organizar e armazenar informação. Compreender estas estruturas é fundamental para manipular e analisar dados eficazmente.

## 1.1. Vetores (Vectors)

Um vetor é a estrutura de dados mais básica em R. É uma sequência de elementos do **mesmo tipo**.

- **Numéricos:** Números inteiros ou decimais.
- **Caracteres (Strings):** Texto.
- **Lógicos (Booleans):** `TRUE` ou `FALSE`.

Podemos criar vetores usando a função `c()` (de "combine" ou "concatenate").

```
# Vetor numérico
idades <- c(18, 19, 17, 20, 18)
print(idades)
class(idades) # Para verificar o tipo

# Vetor de caracteres
nomes <- c("João", "Maria", "Pedro", "Ana")
print(nomes)
class(nomes)

# Vetor lógico
aprovado <- c(TRUE, FALSE, TRUE, TRUE)
print(aprovado)
class(aprovado)

# Aceder a elementos de um vetor (usando [])
print(idades[1]) # Primeiro elemento
print(nomes[c(2, 4)]) # Segundo e quarto elementos
```

**Importante**: Todos os elementos de um vetor devem ser do mesmo tipo. Se tentarmos misturar, o R fará uma "coerção" para o tipo mais flexível (ex: números tornam-se caracteres se misturados com texto).

```
# Exemplo de coerção
vetor_misto <- c(1, "dois", 3)
print(vetor_misto)
class(vetor_misto) # O que aconteceu?
```

## 1.2. Matrizes (Matrices)

Uma matriz é uma coleção bidimensional de elementos do mesmo tipo, organizada em linhas e colunas. Pensem numa tabela onde todas as células contêm números, ou todas contêm texto. Podemos criar matrizes com a função ‘matrix()’.

```
# Criar uma matriz de 2 linhas e 3 colunas
matriz_numeros <- matrix(c(1, 2, 3, 4, 5, 6), nrow = 2, ncol = 3)
print(matriz_numeros)

# Preencher por linha (por defeito é por coluna)
matriz_por_linha <- matrix(c(1, 2, 3, 4, 5, 6), nrow = 2, byrow = TRUE)
print(matriz_por_linha)

# Aceder a elementos de uma matriz [linha, coluna]
print(matriz_numeros[1, 2]) # Elemento da 1ª linha, 2ª coluna
print(matriz_numeros[ , 1]) # Toda a 1ª coluna
print(matriz_numeros[2, ]) # Toda a 2ª linha
```

## 1.3. Arrays (Arrays)

Arrays são extensões das matrizes para mais de duas dimensões. Raramente usados em análises de dados introdutórias, mas útil saber que existem.

```
# Exemplo de um array 2x3x2
array_exemplo <- array(1:12, dim = c(2, 3, 2))
print(array_exemplo)
```

## 1.4. Listas (Lists)

Uma lista é uma coleção de objetos que podem ser de **diferentes tipos** e tamanhos. Pensem numa lista como uma "caixa" que pode conter outras "caixas" (vetores, matrizes, data frames, ou até outras listas). Podemos criar listas com a função `list()`.

```
# Criar uma lista com diferentes tipos de objetos
minha_lista <- list(
  nome = "Carlos",
  idade = 17,
  notas = c(15, 18, 14),
  disciplinas = c("Matemática", "Física"),
  informacao_extra = matrix(c(1,2,3,4), nrow = 2)
)
print(minha_lista)

# Aceder a elementos de uma lista
# Usar `$` para aceder pelo nome
print(minha_lista$nome)
print(minha_lista$notas)

# Usar `[[ ]]` para aceder pelo índice (e manter o tipo do elemento)
print(minha_lista[[3]])
class(minha_lista[[3]])

# Usar `[ ]` para aceder pelo índice (e obter uma sub-lista)
print(minha_lista[3])
class(minha_lista[3]) # Note que o resultado ainda é uma lista!
```

## 1.5. Data Frames (Data Frames) - O Nosso Foco!

Um `data frame` é a estrutura de dados **mais importante** para a Ciência de Dados em R. É uma lista de vetores de igual comprimento, onde cada vetor representa uma coluna e cada elemento do vetor representa uma linha. Pensem num `data frame` como uma **tabela** ou uma **folha de cálculo**, onde: \* Cada **coluna** pode ter um tipo de dados diferente (numérico, texto, lógico, etc.). \* Cada **linha** representa uma observação ou um registo. Podemos criar um `data frame` usando a função `data.frame()`.

```
# Criar um data frame
alunos_df <- data.frame(
  Nome = c("Ana", "Bruno", "Carla", "Diogo"),
  Idade = c(17, 18, 17, 19),
  Altura_cm = c(165, 180, 170, 175),
  Genero = c("F", "M", "F", "M"),
  Nota_Mat = c(16, 14, 18, 15)
)
print(alunos_df)

# Ver a estrutura do data frame
str(alunos_df)

# Ver as dimensões (linhas x colunas)
dim(alunos_df)

# Ver os nomes das colunas
colnames(alunos_df)

# Aceder a colunas (variáveis) de um data frame
print(alunos_df$Nome)
print(alunos_df[["Idade"]])

# Aceder a linhas e colunas [linha, coluna]
print(alunos_df[1, ]) # Primeira linha
print(alunos_df[ , "Altura_cm"]) # Coluna 'Altura_cm'
print(alunos_df[2, 3]) # Elemento da 2ª linha, 3ª coluna
```

Os `data frames` são incrivelmente versáteis e serão a base de quase todas as nossas análises.
