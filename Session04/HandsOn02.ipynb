# 1. Medidas de Dispersão

As medidas de dispersão (ou variabilidade) indicam o quão espalhados os dados estão em relação ao seu centro.
São tão importantes quanto as medidas de tendência central, pois dois conjuntos de dados podem ter a mesma média, mas dispersões muito diferentes.
Vamos usar o dataset `iris` novamente.

```
# Carregar o dataset iris
data(iris)
head(iris)
```

## 1.1. Amplitude (Range)

A amplitude é a diferença entre o valor máximo e o valor mínimo de um conjunto de dados.
É a medida de dispersão mais simples, mas é muito sensível a outliers.
Função em R: `range()` (retorna min e max) e `diff(range())` para a amplitude.

```
# Amplitude do Sepal.Length
amplitude_sepal_length <- diff(range(iris$Sepal.Length))
print(paste("Amplitude do Sepal.Length:", round(amplitude_sepal_length, 2)))

# Mínimo e Máximo do Sepal.Length
min_max_sepal_length <- range(iris$Sepal.Length)
print(paste("Mínimo Sepal.Length:", min_max_sepal_length[1]))
print(paste("Máximo Sepal.Length:", min_max_sepal_length[2]))
```

## 1.2. Variância (Variance)

A variância mede a dispersão média dos pontos de dados em relação à média.
É calculada como a média dos quadrados das diferenças de cada valor em relação à média.
Função em R: `var()`

```
# Variância do Sepal.Length
variancia_sepal_length <- var(iris$Sepal.Length)
print(paste("Variância do Sepal.Length:", round(variancia_sepal_length, 2)))

# Exemplo com NA
dados_com_na <- c(10, 20, 30, NA, 40)
var(dados_com_na, na.rm = TRUE)
```

## 1.3. Desvio Padrão (Standard Deviation)

O desvio padrão é a raiz quadrada da variância.
É a medida de dispersão mais utilizada, pois está na mesma unidade de medida dos dados originais, tornando-a mais fácil de interpretar do que a variância.
Função em R: `sd()`

```
# Desvio Padrão do Sepal.Length
desvio_padrao_sepal_length <- sd(iris$Sepal.Length)
print(paste("Desvio Padrão do Sepal.Length:", round(desvio_padrao_sepal_length, 2)))

# Exemplo com NA
sd(dados_com_na, na.rm = TRUE)
```

## 1.4. Quartis e Intervalo Interquartil (IQR)

Os quartis dividem um conjunto de dados ordenado em quatro partes iguais.
\* **1º Quartil (Q1):** 25% dos dados estão abaixo deste valor.
\* **2º Quartil (Q2):** É a mediana (50% dos dados estão abaixo).
\* **3º Quartil (Q3):** 75% dos dados estão abaixo deste valor.
O Intervalo Interquartil (IQR) é a diferença entre o 3º Quartil e o 1º Quartil (IQR = Q3 - Q1).
Ele representa a amplitude dos 50% centrais dos dados, sendo menos afetado por outliers do que a amplitude total.
Funções em R: `quantile()` e `IQR()`

```
# Quartis do Sepal.Length
quartis_sepal_length <- quantile(iris$Sepal.Length)
print("Quartis do Sepal.Length:")
print(quartis_sepal_length)

# Intervalo Interquartil (IQR) do Sepal.Length
iqr_sepal_length <- IQR(iris$Sepal.Length)
print(paste("IQR do Sepal.Length:", round(iqr_sepal_length, 2)))

# Também podemos especificar percentis específicos
quantile(iris$Sepal.Length, probs = c(0.10, 0.90)) # 10º e 90º percentil
```

## 1.5. Resumo Geral (summary())

A função summary() que já vimos na Sessão 2 é excelente porque fornece várias destas medidas de uma só vez para cada variável numérica (Mínimo, Q1, Mediana, Média, Q3, Máximo).

```
summary(iris$Sepal.Length)
```
