# 1. Medidas de Tendência Central

As medidas de tendência central são valores que representam o "centro" ou o "valor típico" de um conjunto de dados. São fundamentais para resumir e entender a distribuição de uma variável.

Vamos usar o dataset `iris` para demonstrar estas medidas.

```
# Carregar o dataset iris
data(iris)
head(iris)
```

## 1.1. Média (Mean)

A média é a soma de todos os valores dividida pelo número de valores. É a medida de tendência central mais comum, mas é sensível a valores extremos (outliers). Função em R: `mean()`

```
# Média do Sepal.Length
media_sepal_length <- mean(iris$Sepal.Length)
print(paste("Média do Sepal.Length:", round(media_sepal_length, 2)))

# Média do Petal.Width
media_petal_width <- mean(iris$Petal.Width)
print(paste("Média do Petal.Width:", round(media_petal_width, 2)))

# Exemplo com NA (Not Available)
dados_com_na <- c(10, 20, 30, NA, 40)
mean(dados_com_na) # Retorna NA por defeito
mean(dados_com_na, na.rm = TRUE) # Ignora os NAs e calcula a média
```

## 1.2. Mediana (Median)

A mediana é o valor do meio de um conjunto de dados quando estes estão ordenados. Se houver um número par de observações, é a média dos dois valores centrais. É menos sensível a outliers do que a média. Função em R: `median()`

```
# Mediana do Sepal.Length
mediana_sepal_length <- median(iris$Sepal.Length)
print(paste("Mediana do Sepal.Length:", round(mediana_sepal_length, 2)))

# Mediana do Petal.Width
mediana_petal_width <- median(iris$Petal.Width)
print(paste("Mediana do Petal.Width:", round(mediana_petal_width, 2)))

# Exemplo com NA
median(dados_com_na, na.rm = TRUE)
```

## 1.3. Moda (Mode)

A moda é o valor que aparece com maior frequência num conjunto de dados. Pode haver uma moda (unimodal), várias modas (multimodal) ou nenhuma moda. O R não tem uma função mode() embutida para calcular a moda de variáveis numéricas ou categóricas de forma geral (a função mode() em R tem outro significado, relacionado com o tipo de armazenamento de um objeto). Precisamos de criar uma função ou usar pacotes. Vamos criar uma função simples para calcular a moda:

```
# Função para calcular a moda
get_mode <- function(v) {
  uniqv <- unique(v)
  # tabulate conta a frequência de cada valor único
  # which.max encontra o índice do valor mais frequente
  # uniqv[...] retorna o valor mais frequente
  uniqv[which.max(tabulate(match(v, uniqv)))]
}

# Moda da coluna Species (categórica)
moda_species <- get_mode(iris$Species)
print(paste("Moda da Espécie:", moda_species))

# Moda da coluna Sepal.Length (numérica - pode não ser muito informativa se houver muitos valores únicos)
moda_sepal_length <- get_mode(iris$Sepal.Length)
print(paste("Moda do Sepal.Length:", moda_sepal_length))

# Para variáveis categóricas, podemos usar table()
table(iris$Species)
```

## 1.4. Quando Usar Cada Medida?

- **Média:** Ideal para dados simetricamente distribuídos sem outliers.
- **Mediana:** Melhor para dados assimétricos ou com outliers, pois não é afetada por valores extremos.
- **Moda:** Útil para variáveis categóricas ou para identificar o valor mais comum em variáveis numéricas discretas.

**Parabéns!** Já dominam as medidas de tendência central!
