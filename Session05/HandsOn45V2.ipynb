Variáveis chave no dataset diamonds:

* price: preço em dólares (numérico)
* carat: peso do diamante (numérico)
* cut: qualidade do corte (categórica: Fair, Good, Very Good, Premium, Ideal)
* color: cor do diamante (categórica: de D (melhor) a J (pior))
* clarity: clareza do diamante (categórica: de I1 (pior) a IF (melhor))
* depth: profundidade total % (numérico)
* table: largura do topo do diamante em relação ao ponto mais largo (numérico)
* x, y, z: dimensões em mm (numéricas)
